### Q1. For 'House of Arrest', what is the show id and Who is the director of this show?

### isin()

```python

data=pd.read_csv("file.csv")
data=pd.DataFrame(data)

data[data["Title"].isin(["House of Cards"])]
# data.head()
```

```python
# To show all the records of a particular item  in any column

# data["Title"].isin(["House of cards"])

# Filter rows where the Title is 'House of cards'
data[data['Title'].isin(["House Arrest"])]
```

### str.contains()

```python
#To show all the records of a particular string in any column
# data['Title'].str.contains("House of cards")
data[data['Title'].str.contains("House Arrest")]
```

### Q3. How many movies and TV shows are in the dataset? show in bar graph

### Groupby()

```python
# To show all the unique items in columns and show their count
data.groupby("Category").Category.count()
```

### countplot()

```python
# to show the count of all unique values of any column in the form of a graph 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# sns.countplot(data["Category"])
# Calculate the counts of unique values
value_counts = data['Category'].value_counts()
print(value_counts)

# # Plot the counts using a bar plot
# plt.figure(figsize=(8, 6))
sns.barplot(x=value_counts.index, y=value_counts.values)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Count of Unique Values in Category Column')
plt.show()

# sns.countplot(data["Category"])
```
